## 3) Frequent Itemset Mining with FP‑Growth (min_support = 0.02) (20 pts)
We pivot the data to a **transaction × item** one‑hot table (boolean), then run FP‑Growth.

```python
freq_fp = fp_growth(df, min_support=0.02, use_colnames=True)
freq_fp
df = pd.read_csv(bread_path)
bread_path = "bread_basket.csv"
```

## 4) Association Rules + Report Table (30 pts)
(metric = confidence, min_threshold = ?) Please find a suitable min_threshold

```python
rules = association_rules(freq_fp, metric='confidence', min_threshold=0.1)
print(rules[['antecedents', 'consequences', 'support', 'confidence', 'lift']].sort_values('confidence', ascending=False))

if rules.empty: print("No rules found")
```

## 5) Interpretation (10 pts)
**Interpret the rule `{Coffee, Cake} ⇒ {Bread}` in plain English.**

- **Support**: What fraction of *all* transactions contain Coffee, Cake, and Bread together?
- **Confidence**: Among baskets with Coffee and Cake, what share also include Bread?
- **Lift > 1** implies positive association; comment on practical meaning.

This rule explains that customers buying coffee and cake are likely to buy bread as well.
Confidence = coffee, cake, bread / coffee, cake
Coffee and cake buyers are more like to buy bread and can be offered as a bundle. Lift being greater than 1 indicates that there is a positive association.
